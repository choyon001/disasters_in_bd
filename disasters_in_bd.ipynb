# Most affected district by tropical cyclone

```python
districts_list = [
    "Dhaka", "Faridpur", "Gazipur", "Gopalganj", "Jamalpur", "Kishoreganj", 
    "Madaripur", "Manikganj", "Munshiganj", "Mymensingh", "Narayanganj", "Narsingdi", 
    "Netrokona", "Rajbari", "Shariatpur", "Sherpur", "Tangail", "Bogra", "Joypurhat", 
    "Naogaon", "Natore", "Nawabganj", "Pabna", "Rajshahi", "Sirajgonj", "Dinajpur", 
    "Gaibandha", "Kurigram", "Lalmonirhat", "Nilphamari", "Panchagarh", "Rangpur", 
    "Thakurgaon", "Barguna", "Barisal", "Bhola", "Jhalokati", "Patuakhali", "Pirojpur", 
    "Bandarban", "Brahmanbaria", "Chandpur", "Chittagong", "Comilla", "Cox's Bazar", 
    "Feni", "Khagrachari", "Lakshmipur", "Noakhali", "Rangamati", "Habiganj", 
    "Maulvibazar", "Sunamganj", "Sylhet", "Bagerhat", "Chuadanga", "Jessore", 
    "Jhenaidah", "Khulna", "Kushtia", "Magura", "Meherpur", "Narail", "Satkhira"
]
district_dict = {}
for i in districts_list:
    district_dict[i]=0
```

```python
Tropical_cyclone=data[data['Disaster Subtype']=='Tropical cyclone']
locations = Tropical_cyclone['Geo Locations'].str.split(',')
locations_list=[]
for location in locations:
    for i in location:
        if i not in locations_list:
            locations_list.append(i.strip())

            
```

```python
for location in locations_list :
    for district in districts_list:
        if district in location:
            district_dict[district]+=1
```

```python
# Extract data for plotting
locations = list(district_dict.keys())
values = list(district_dict.values())

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(locations, values, color='skyblue')
plt.xlabel('Districts')
plt.ylabel('Number of cyclone')
plt.title(' Most affected  Districts by Tropical Cyclone')
plt.xticks(rotation=90)  # Rotate x-labels for better readability
plt.tight_layout()

# Show the chart
plt.show()
```

# Most affted area by cold wave(extreme temperature) from location

```python
cold_wave_districts = {}
for i in districts_list:
    cold_wave_districts[i]=0
    
```

```python
cold_wave_list=data[data['Disaster Type']=='Extreme temperature']
locations = cold_wave_list['Location'].str.split(',')
locations_list=[]
for location in locations:
    try:
        for i in location:
            if i not in locations_list:
                locations_list.append(i.strip())
    except:
        pass
```

```python
dataa = [entry.replace('(', ',') for entry in locations_list]
ans = []
# Print the updated list
for entry in dataa:
    one = entry.split(',')
    for i in one:
        if 'province' in i:
            one.remove(i)
    ans.append(one)
```

```python
for i in ans:
    for j in i:
        for district in districts_list:
            if district in j :
                cold_wave_districts[district]+=1
```

```python
# Extract data for plotting
locations = list(cold_wave_districts.keys())
values = list(cold_wave_districts.values())

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(locations, values, color='skyblue')
plt.xlabel('Districts')
plt.ylabel('Number of cold wave')
plt.title(' Most affected  Districts by cold wave')
plt.xticks(rotation=90)  # Rotate x-labels for better readability
plt.tight_layout()

# Show the chart
plt.show()
```
